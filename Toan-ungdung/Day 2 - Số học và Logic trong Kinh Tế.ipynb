## Phần 2: Vương Quốc Kinh Doanh Bất Động Sản

> Tưởng tượng bạn đang ngồi trong văn phòng của một công ty đầu tư lớn, với bản đồ các dự án bất động sản và chứng khoán trải rộng trên bàn. Mỗi dự án là một cơ hội, và công cụ bạn cần để đánh giá chúng chính là Python, với "biến" và "kiểu dữ liệu" là những chìa khóa.

### 1. Biến - Những Khoản Đầu Tư:

Bạn có một dự án bất động sản ở trung tâm thành phố và một cơ hội đầu tư vào thị trường chứng khoán đang lên. Hãy đặt tên cho chúng:

> Biến ở đây giống như các khoản đầu tư bạn sẵn sàng rót vào mỗi dự án, mỗi con số đều chứa đựng một câu chuyện về hy vọng và rủi ro.Chúng ta đang sử dụng hai biến để đại diện cho hai khoản đầu tư khác nhau: `dauTu_BatDongSan` và `dauTu_ChungKhoan`. Mỗi biến này đều lưu trữ một giá trị số, và kiểu dữ liệu của chúng là số nguyên `(int)`.

> Giải thích Biến:
Biến: Trong lập trình, biến là một cái tên được gán cho một vị trí trong bộ nhớ, nơi có thể lưu trữ dữ liệu. Biến có thể được sử dụng để lưu trữ thông tin và thay đổi giá trị thông qua quá trình thực thi chương trình.


```python
dauTu_BatDongSan = 5000000 # 5 triệu cho dự án bất động sản
dauTu_ChungKhoan = 2000000 # 2 triệu cho cơ hội chứng khoán
```

#### Giải thích Kiểu Dữ liệu:

> **Kiểu Dữ liệu (Data Type):** Kiểu dữ liệu xác định loại giá trị mà một biến có thể lưu trữ, ví dụ như số, văn bản, danh sách, v.v. Trong Python, kiểu dữ liệu cơ bản bao gồm int (số nguyên), float (số thực), str (chuỗi ký tự), và boolean (giá trị đúng/sai).


## Tính Toán - Vẽ Bức Tranh Lợi Nhuận: Khi đã có bản đồ và la bàn, hãy bắt đầu hành trình tìm kiếm kho báu:

<center>
<img src="https://i.gifer.com/KCIo.gif" />
</center>

```python
tyLeLoiNhuan_BatDongSan = 0.1 # Mong đợi 10% lợi nhuận từ bất động sản
tyLeLoiNhuan_ChungKhoan = 0.15 # Mong đợi 15% lợi nhuận từ chứng khoán

loiNhuan_BatDongSan = dauTu_BatDongSan * tyLeLoiNhuan_BatDongSan
loiNhuan_ChungKhoan = dauTu_ChungKhoan * tyLeLoiNhuan_ChungKhoan

print("Vốn ban đầu khi đầu tư Bất động sản:", dauTu_BatDongSan)
print("Vốn ban đầu khi đầu tư Chứng khoán:", dauTu_ChungKhoan)
print("-------------------------------------")
print("Lợi nhuận từ Bất động sản:", loiNhuan_BatDongSan)
print("Lợi nhuận từ Chứng khoán:", loiNhuan_ChungKhoan)
```

> **Ý Nghĩa:** Qua đoạn code này, bạn không chỉ thực hiện được các phép tính toán đơn giản trong Python mà còn có thể áp dụng kiến thức này vào việc đánh giá các cơ hội đầu tư trong thực tế. Việc tính toán này giống như việc vẽ nên bức tranh tương lai với những gam màu lợi nhuận và hy vọng, giúp bạn hình dung được kết quả mà mỗi quyết định đầu tư có thể mang lại, từ đó đưa ra lựa chọn đầu tư sáng suốt và hiệu quả.

> Hãy bước vào khu vườn của những ước mơ đầu tư, nơi mỗi khoản đầu tư giống như một hạt giống đầy hứa hẹn. Chúng ta sẽ cùng nhau trải qua một hành trình chăm sóc và nuôi dưỡng chúng, với sự hỗ trợ của Python, như một người bạn đồng hành đáng tin cậy.

* Bước 1: Chuẩn Bị Mảnh Đất Mơ Ước
* Bước 2: Chăm Sóc với Nước và Phân Bón
* Bước 3: Theo Dõi Sự Phát Triển
* Bước 4: Thu Hoạch



```python
# Giả sử ban đầu bạn có một số vốn để đầu tư
von_dau_tu = 10000  # 10.000 đô la

# Tỷ lệ lợi nhuận hàng năm dự kiến từ trái phiếu chính phủ và cổ phiếu công ty lớn
ty_le_loi_nhuan_mong_muon = 0.05  # 5% mỗi năm

# Tính toán lợi nhuận sau 5 năm
nam = 5
loi_nhuan = von_dau_tu * (1 + ty_le_loi_nhuan_mong_muon) ** nam - von_dau_tu

print("Sau", nam, "năm, lợi nhuận từ vườn ươm đầu tư của bạn sẽ là:", round(loi_nhuan, 2), "đô la.")
```

```python
# @title Khám phá lợi nhuận khu vườn thay đổi sau hằng năm { run: "auto" }
nam = 7 # @param {type:"slider", min:1, max:50, step:1}

loi_nhuan = von_dau_tu * (1 + ty_le_loi_nhuan_mong_muon) ** nam - von_dau_tu
print(f'Sau {nam} năm, lợi nhuận từ vườn ươm đầu tư của bạn sẽ là: {round(loi_nhuan, 2):,} Đô la.')
```

##🚀 Kịch Bản "Chinh Phục Mặt Trăng": Tăng Trưởng Nhanh
Hãy cùng nhau bước lên tàu vũ trụ của những nhà đầu tư táo bạo, chuẩn bị cho chuyến "Chinh Phục Mặt Trăng" đầy kỳ thú. Đây không phải là cuộc phiêu lưu bình thường, mà là hành trình đầu tư vào những lĩnh vực mới mẻ như công nghệ tiên tiến hoặc thị trường mới nổi. Có thể có những rủi ro không lường trước được nhưng cũng chứa đựng cơ hội tăng trưởng phi thường.

<center>
  <img height="320px" width="100%" src="https://www.icegif.com/wp-content/uploads/icegif-590.gif"/>
</center>

```python
# 🌟 Bước 1: Chuẩn Bị Tàu Vũ Trụ
# Lên kế hoạch: Xác định số tiền đầu tư cho chuyến phiêu lưu này
khoanDauTu = 10000 # Đặt số tiền đầu tư là 10.000 đô la

# 🌟 Bước 2: Định Vị Hành Tinh Mục Tiêu
# Phân tích điểm đến: Xác định tỷ lệ lợi nhuận dự kiến và mức rủi ro tiềm ẩn
tyLeLoiNhuan = 0.2 # Kỳ vọng lợi nhuận 20%
mucRuiRo = 0.1 # Rủi ro tiềm ẩn 10%

# 🌟 Bước 3: Phóng Tàu và Theo Dõi Hành Trình
# Dự đoán hành trình: Tính toán lợi nhuận và rủi ro dự kiến từ chuyến đầu tư
loiNhuanDuKien = khoanDauTu * tyLeLoiNhuan # Tính lợi nhuận dự kiến
ruiRoDuKien = khoanDauTu * mucRuiRo # Tính rủi ro tiềm ẩn

# 🌟 Bước 4: Đánh Giá Kết Quả Hành Trình
# Phân tích kết quả: Đánh giá và hiểu biết kết quả dự kiến từ chuyến phiêu lưu
print(f"Lợi nhuận dự kiến từ chuyến du hành: {loiNhuanDuKien} đô la.")
print(f"Rủi ro có thể gặp phải: {ruiRoDuKien} đô la.")
```

> Qua từng bước, từ việc lên kế hoạch đầu tư cho đến đánh giá kết quả dự kiến, Python giúp chúng ta minh họa rõ ràng mỗi phần của cuộc phiêu lưu "Chinh Phục Mặt Trăng". Đoạn mã này không chỉ giúp bạn tính toán lợi nhuận và rủi ro tiềm ẩn, mà còn mở ra cánh cửa cho sự mạo hiểm đầy thông minh và có tính toán.


##🧩 Kịch Bản "Bức Tranh": Đa Dạng Hóa
> Hãy tưởng tượng bạn đang đứng trước một tờ giấy trắng lớn, trong tay là bảng màu của những cơ hội đầu tư. Mỗi lựa chọn của bạn, từ bất động sản, cổ phiếu, đến trái phiếu, giống như việc chọn một màu sắc khác nhau để phác họa lên tờ giấy ấy. Hãy hình dung bạn là một họa sĩ tài ba, đang đứng trước một bức tranh còn trống rỗng, sẵn sàng để khắc họa những ước mơ và tham vọng của mình. Trong tay bạn là bảng màu đầy ắp những lựa chọn đầu tư: `dauTu_TaiSanA`, `dauTu_TaiSanB`, `dauTu_TaiSanC`... Mỗi khoản đầu tư giống như một màu sơn đặc trưng, từ màu xanh lá của sự an toàn, bền vững, đến màu đỏ rực rỡ của sự đam mê, mạo hiểm, và cơ hội lớn.

<center>
<img height="200px" src="https://discovertemplate.com/wp-content/uploads/2020/09/DT_G33_Finance-Animated-GIF-Icon-pack.gif"/>
</center>

* `dauTu_TaiSanA` có thể là một bức tranh về bất động sản, vững chắc và kiên cố, mang màu xanh của sự an nhàn và ổn định. `dauTu_TaiSanB` lại như những cổ phiếu công nghệ mới nổi, đầy màu sắc và sự biến đổi, chói lọi với màu vàng của sự sáng tạo và đổi mới. Trong khi đó, `dauTu_TaiSanC` có thể là trái phiếu chính phủ, với màu xám của sự đáng tin cậy và an toàn.

* Khi bạn kết hợp những mảnh ghép này, không chỉ đơn thuần là lựa chọn giữa màu sắc, mà là việc tạo ra một bức tranh đầu tư đa dạng, cân đối giữa rủi ro và cơ hội. Mỗi nét vẽ, mỗi lựa chọn, đều là một phần của chiến lược đa dạng hóa, nhằm giảm thiểu rủi ro tổng thể và tối đa hóa cơ hội tăng trưởng.

Dựa vào câu chuyện về việc đa dạng hóa đầu tư như một bức tranh đầy màu sắc, chúng ta có thể mô phỏng qua một đoạn mã Python để thể hiện việc phân bổ vốn vào các loại tài sản khác nhau.

Mỗi loại tài sản sẽ được đại diện bằng một biến, và chúng ta sẽ sử dụng tỷ lệ phần trăm của tổng số vốn đầu tư cho mỗi loại để thể hiện việc phân chia đầu tư.


```python
# Tổng số tiền đầu tư
tongVonDauTu = 10000  # Giả sử bạn có 10.000 đô la để đầu tư

# Định nghĩa tỷ lệ phần trăm đầu tư cho mỗi loại tài sản
tyLe_TaiSanA = 0.4  # 40% cho Tài sản A - có thể là bất động sản
tyLe_TaiSanB = 0.3  # 30% cho Tài sản B - có thể là cổ phiếu công nghệ
tyLe_TaiSanC = 0.3  # 30% cho Tài sản C - có thể là trái phiếu hoặc quỹ đầu tư an toàn

# Tính toán số tiền đầu tư cho mỗi loại tài sản
dauTu_TaiSanA = tongVonDauTu * tyLe_TaiSanA
dauTu_TaiSanB = tongVonDauTu * tyLe_TaiSanB
dauTu_TaiSanC = tongVonDauTu * tyLe_TaiSanC

# In ra số tiền đầu tư cho mỗi loại tài sản
print(f"Số tiền đầu tư vào Tài sản A: {dauTu_TaiSanA} đô la")
print(f"Số tiền đầu tư vào Tài sản B: {dauTu_TaiSanB} đô la")
print(f"Số tiền đầu tư vào Tài sản C: {dauTu_TaiSanC} đô la")

# In ra tổng quan về bức tranh đầu tư
print("\nBức tranh đầu tư của bạn bao gồm:")
print(f"- Màu xanh của sự an nhàn từ Tài sản A với {tyLe_TaiSanA*100}% tổng vốn.")
print(f"- Sắc đỏ của cơ hội và mạo hiểm từ Tài sản B với {tyLe_TaiSanB*100}% tổng vốn.")
print(f"- Màu vàng của sự cân bằng từ Tài sản C với {tyLe_TaiSanC*100}% tổng vốn.")
```

> Dưới đây là đoạn code Python hoàn chỉnh cho kịch bản "Suối Nguồn", thể hiện việc tạo ra và tận hưởng thu nhập thụ động từ việc đầu tư:





```python
# Bước 1: Chọn Mảnh Đất Màu Mỡ
# Định Vị Suối Nguồn: Xác định số tiền đầu tư ban đầu
dauTuBanDau = 50000  # 50.000 đô la là số tiền bạn đầu tư ban đầu

# Bước 2: Nuôi Dưỡng Suối Nguồn
# Lên Kế Hoạch Dòng Chảy: Xác định tỷ lệ thu nhập thụ động mong đợi từ đầu tư
thuNhapThuDong = 0.05  # Kỳ vọng 5% thu nhập thụ động hàng năm từ tổng số vốn đầu tư

# Bước 3: Tận Hưởng Dòng Chảy Thu Nhập
# Tính toán số tiền thu nhập thụ động hàng năm
thuNhapHangNam = dauTuBanDau * thuNhapThuDong
print(f"Mỗi năm, suối nguồn đầu tư của bạn sẽ mang lại {thuNhapHangNam:.2f} đô la dưới dạng thu nhập thụ động.")

# Bước 4: Đánh Giá và Mở Rộng Suối Nguồn
# Xem xét cơ hội đầu tư mới hoặc cải thiện suối nguồn hiện tại để tăng thu nhập thụ động
print("""Hãy xem xét cơ hội đầu tư mới hoặc tìm cách cải thiện và
mở rộng suối nguồn thu nhập thụ động của bạn để nó mang lại nhiều giá trị hơn cho cuộc sống của bạn.""")
```
