### 1. Biáº¿n - Nhá»¯ng Khoáº£n Äáº§u TÆ°:

Báº¡n cÃ³ má»™t dá»± Ã¡n báº¥t Ä‘á»™ng sáº£n á»Ÿ trung tÃ¢m thÃ nh phá»‘ vÃ  má»™t cÆ¡ há»™i Ä‘áº§u tÆ° vÃ o thá»‹ trÆ°á»ng chá»©ng khoÃ¡n Ä‘ang lÃªn. HÃ£y Ä‘áº·t tÃªn cho chÃºng:

Má»—i biáº¿n nÃ y Ä‘á»u lÆ°u trá»¯ má»™t giÃ¡ trá»‹ sá»‘, vÃ  kiá»ƒu dá»¯ liá»‡u cá»§a chÃºng lÃ  sá»‘ nguyÃªn `(int)`.

> Giáº£i thÃ­ch Biáº¿n:
Biáº¿n: Trong láº­p trÃ¬nh, biáº¿n lÃ  má»™t cÃ¡i tÃªn Ä‘Æ°á»£c gÃ¡n cho má»™t vá»‹ trÃ­ trong bá»™ nhá»›, nÆ¡i cÃ³ thá»ƒ lÆ°u trá»¯ dá»¯ liá»‡u. Biáº¿n cÃ³ thá»ƒ Ä‘Æ°á»£c sá»­ dá»¥ng Ä‘á»ƒ lÆ°u trá»¯ thÃ´ng tin vÃ  thay Ä‘á»•i giÃ¡ trá»‹ thÃ´ng qua quÃ¡ trÃ¬nh thá»±c thi chÆ°Æ¡ng trÃ¬nh.


```python
dauTu_BatDongSan = 5000000 # 5 triá»‡u cho dá»± Ã¡n báº¥t Ä‘á»™ng sáº£n
dauTu_ChungKhoan = 2000000 # 2 triá»‡u cho cÆ¡ há»™i chá»©ng khoÃ¡n
```

#### Giáº£i thÃ­ch Kiá»ƒu Dá»¯ liá»‡u:

> **Kiá»ƒu Dá»¯ liá»‡u (Data Type):** Kiá»ƒu dá»¯ liá»‡u xÃ¡c Ä‘á»‹nh loáº¡i giÃ¡ trá»‹ mÃ  má»™t biáº¿n cÃ³ thá»ƒ lÆ°u trá»¯, vÃ­ dá»¥ nhÆ° sá»‘, vÄƒn báº£n, danh sÃ¡ch, v.v. ## TÃ­nh ToÃ¡n - Váº½ Bá»©c Tranh Lá»£i Nhuáº­n: Khi Ä‘Ã£ cÃ³ báº£n Ä‘á»“ vÃ  la bÃ n, hÃ£y báº¯t Ä‘áº§u hÃ nh trÃ¬nh tÃ¬m kiáº¿m kho bÃ¡u:

```python
tyLeLoiNhuan_BatDongSan = 0.1 # Mong Ä‘á»£i 10% lá»£i nhuáº­n tá»« báº¥t Ä‘á»™ng sáº£n
tyLeLoiNhuan_ChungKhoan = 0.15 # Mong Ä‘á»£i 15% lá»£i nhuáº­n tá»« chá»©ng khoÃ¡n

loiNhuan_BatDongSan = dauTu_BatDongSan * tyLeLoiNhuan_BatDongSan
loiNhuan_ChungKhoan = dauTu_ChungKhoan * tyLeLoiNhuan_ChungKhoan

print("Vá»‘n ban Ä‘áº§u khi Ä‘áº§u tÆ° Báº¥t Ä‘á»™ng sáº£n:", dauTu_BatDongSan)
print("Vá»‘n ban Ä‘áº§u khi Ä‘áº§u tÆ° Chá»©ng khoÃ¡n:", dauTu_ChungKhoan)
print("-------------------------------------")
print("Lá»£i nhuáº­n tá»« Báº¥t Ä‘á»™ng sáº£n:", loiNhuan_BatDongSan)
print("Lá»£i nhuáº­n tá»« Chá»©ng khoÃ¡n:", loiNhuan_ChungKhoan)
```

> **Ã NghÄ©a:** Qua Ä‘oáº¡n code nÃ y, báº¡n khÃ´ng chá»‰ thá»±c hiá»‡n Ä‘Æ°á»£c cÃ¡c phÃ©p tÃ­nh toÃ¡n Ä‘Æ¡n giáº£n trong Python mÃ  cÃ²n cÃ³ thá»ƒ Ã¡p dá»¥ng kiáº¿n thá»©c nÃ y vÃ o viá»‡c Ä‘Ã¡nh giÃ¡ cÃ¡c cÆ¡ há»™i Ä‘áº§u tÆ° trong thá»±c táº¿. Viá»‡c tÃ­nh toÃ¡n nÃ y giá»‘ng nhÆ° viá»‡c váº½ nÃªn bá»©c tranh tÆ°Æ¡ng lai vá»›i nhá»¯ng gam mÃ u lá»£i nhuáº­n vÃ  hy vá»ng, giÃºp báº¡n hÃ¬nh dung Ä‘Æ°á»£c káº¿t quáº£ mÃ  má»—i quyáº¿t Ä‘á»‹nh Ä‘áº§u tÆ° cÃ³ thá»ƒ mang láº¡i, tá»« Ä‘Ã³ Ä‘Æ°a ra lá»±a chá»n Ä‘áº§u tÆ° sÃ¡ng suá»‘t vÃ  hiá»‡u quáº£.


#HÃ nh TrÃ¬nh Thá»±c Chiáº¿n: Ãp Dá»¥ng Kiáº¿n Thá»©c vÃ o Äáº§u TÆ°

> HÃ£y bÆ°á»›c vÃ o khu vÆ°á»n cá»§a nhá»¯ng Æ°á»›c mÆ¡ Ä‘áº§u tÆ°, nÆ¡i má»—i khoáº£n Ä‘áº§u tÆ° giá»‘ng nhÆ° má»™t háº¡t giá»‘ng Ä‘áº§y há»©a háº¹n. ChÃºng ta sáº½ cÃ¹ng nhau tráº£i qua má»™t hÃ nh trÃ¬nh chÄƒm sÃ³c vÃ  nuÃ´i dÆ°á»¡ng chÃºng, vá»›i sá»± há»— trá»£ cá»§a Python, nhÆ° má»™t ngÆ°á»i báº¡n Ä‘á»“ng hÃ nh Ä‘Ã¡ng tin cáº­y.

* BÆ°á»›c 1: Chuáº©n Bá»‹ Máº£nh Äáº¥t MÆ¡ Æ¯á»›c
* BÆ°á»›c 2: ChÄƒm SÃ³c vá»›i NÆ°á»›c vÃ  PhÃ¢n BÃ³n
* BÆ°á»›c 3: Theo DÃµi Sá»± PhÃ¡t Triá»ƒn
* BÆ°á»›c 4: Thu Hoáº¡ch



```python
# Giáº£ sá»­ ban Ä‘áº§u báº¡n cÃ³ má»™t sá»‘ vá»‘n Ä‘á»ƒ Ä‘áº§u tÆ°
von_dau_tu = 10000  # 10.000 Ä‘Ã´ la

# Tá»· lá»‡ lá»£i nhuáº­n hÃ ng nÄƒm dá»± kiáº¿n tá»« trÃ¡i phiáº¿u chÃ­nh phá»§ vÃ  cá»• phiáº¿u cÃ´ng ty lá»›n
ty_le_loi_nhuan_mong_muon = 0.05  # 5% má»—i nÄƒm

# TÃ­nh toÃ¡n lá»£i nhuáº­n sau 5 nÄƒm
nam = 5
loi_nhuan = von_dau_tu * (1 + ty_le_loi_nhuan_mong_muon) ** nam - von_dau_tu

print("Sau", nam, "nÄƒm, lá»£i nhuáº­n tá»« vÆ°á»n Æ°Æ¡m Ä‘áº§u tÆ° cá»§a báº¡n sáº½ lÃ :", round(loi_nhuan, 2), "Ä‘Ã´ la.")
```

```python
# @title KhÃ¡m phÃ¡ lá»£i nhuáº­n khu vÆ°á»n thay Ä‘á»•i sau háº±ng nÄƒm { run: "auto" }
nam = 7 # @param {type:"slider", min:1, max:50, step:1}

loi_nhuan = von_dau_tu * (1 + ty_le_loi_nhuan_mong_muon) ** nam - von_dau_tu
print(f'Sau {nam} nÄƒm, lá»£i nhuáº­n tá»« vÆ°á»n Æ°Æ¡m Ä‘áº§u tÆ° cá»§a báº¡n sáº½ lÃ : {round(loi_nhuan, 2):,} ÄÃ´ la.')
```

##ğŸš€ Ká»‹ch Báº£n "Chinh Phá»¥c Máº·t TrÄƒng": TÄƒng TrÆ°á»Ÿng Nhanh
HÃ£y cÃ¹ng nhau bÆ°á»›c lÃªn tÃ u vÅ© trá»¥ cá»§a nhá»¯ng nhÃ  Ä‘áº§u tÆ° tÃ¡o báº¡o, chuáº©n bá»‹ cho chuyáº¿n "Chinh Phá»¥c Máº·t TrÄƒng" Ä‘áº§y ká»³ thÃº. ÄÃ¢y khÃ´ng pháº£i lÃ  cuá»™c phiÃªu lÆ°u bÃ¬nh thÆ°á»ng, mÃ  lÃ  hÃ nh trÃ¬nh Ä‘áº§u tÆ° vÃ o nhá»¯ng lÄ©nh vá»±c má»›i máº» nhÆ° cÃ´ng nghá»‡ tiÃªn tiáº¿n hoáº·c thá»‹ trÆ°á»ng má»›i ná»•i. CÃ³ thá»ƒ cÃ³ nhá»¯ng rá»§i ro khÃ´ng lÆ°á»ng trÆ°á»›c Ä‘Æ°á»£c nhÆ°ng cÅ©ng chá»©a Ä‘á»±ng cÆ¡ há»™i tÄƒng trÆ°á»Ÿng phi thÆ°á»ng.

```python
# ğŸŒŸ BÆ°á»›c 1: Chuáº©n Bá»‹ TÃ u VÅ© Trá»¥
# LÃªn káº¿ hoáº¡ch: XÃ¡c Ä‘á»‹nh sá»‘ tiá»n Ä‘áº§u tÆ° cho chuyáº¿n phiÃªu lÆ°u nÃ y
khoanDauTu = 10000 # Äáº·t sá»‘ tiá»n Ä‘áº§u tÆ° lÃ  10.000 Ä‘Ã´ la

# ğŸŒŸ BÆ°á»›c 2: Äá»‹nh Vá»‹ HÃ nh Tinh Má»¥c TiÃªu
# PhÃ¢n tÃ­ch Ä‘iá»ƒm Ä‘áº¿n: XÃ¡c Ä‘á»‹nh tá»· lá»‡ lá»£i nhuáº­n dá»± kiáº¿n vÃ  má»©c rá»§i ro tiá»m áº©n
tyLeLoiNhuan = 0.2 # Ká»³ vá»ng lá»£i nhuáº­n 20%
mucRuiRo = 0.1 # Rá»§i ro tiá»m áº©n 10%

# ğŸŒŸ BÆ°á»›c 3: PhÃ³ng TÃ u vÃ  Theo DÃµi HÃ nh TrÃ¬nh
# Dá»± Ä‘oÃ¡n hÃ nh trÃ¬nh: TÃ­nh toÃ¡n lá»£i nhuáº­n vÃ  rá»§i ro dá»± kiáº¿n tá»« chuyáº¿n Ä‘áº§u tÆ°
loiNhuanDuKien = khoanDauTu * tyLeLoiNhuan # TÃ­nh lá»£i nhuáº­n dá»± kiáº¿n
ruiRoDuKien = khoanDauTu * mucRuiRo # TÃ­nh rá»§i ro tiá»m áº©n

# ğŸŒŸ BÆ°á»›c 4: ÄÃ¡nh GiÃ¡ Káº¿t Quáº£ HÃ nh TrÃ¬nh
# PhÃ¢n tÃ­ch káº¿t quáº£: ÄÃ¡nh giÃ¡ vÃ  hiá»ƒu biáº¿t káº¿t quáº£ dá»± kiáº¿n tá»« chuyáº¿n phiÃªu lÆ°u
print(f"Lá»£i nhuáº­n dá»± kiáº¿n tá»« chuyáº¿n du hÃ nh: {loiNhuanDuKien} Ä‘Ã´ la.")
print(f"Rá»§i ro cÃ³ thá»ƒ gáº·p pháº£i: {ruiRoDuKien} Ä‘Ã´ la.")
```

> Qua tá»«ng bÆ°á»›c, tá»« viá»‡c lÃªn káº¿ hoáº¡ch Ä‘áº§u tÆ° cho Ä‘áº¿n Ä‘Ã¡nh giÃ¡ káº¿t quáº£ dá»± kiáº¿n, Python giÃºp chÃºng ta minh há»a rÃµ rÃ ng má»—i pháº§n cá»§a cuá»™c phiÃªu lÆ°u "Chinh Phá»¥c Máº·t TrÄƒng". Dá»±a vÃ o cÃ¢u chuyá»‡n vá» viá»‡c Ä‘a dáº¡ng hÃ³a Ä‘áº§u tÆ° nhÆ° má»™t bá»©c tranh Ä‘áº§y mÃ u sáº¯c, chÃºng ta cÃ³ thá»ƒ mÃ´ phá»ng qua má»™t Ä‘oáº¡n mÃ£ Python Ä‘á»ƒ thá»ƒ hiá»‡n viá»‡c phÃ¢n bá»• vá»‘n vÃ o cÃ¡c loáº¡i tÃ i sáº£n khÃ¡c nhau.

Má»—i loáº¡i tÃ i sáº£n sáº½ Ä‘Æ°á»£c Ä‘áº¡i diá»‡n báº±ng má»™t biáº¿n, vÃ  chÃºng ta sáº½ sá»­ dá»¥ng tá»· lá»‡ pháº§n trÄƒm cá»§a tá»•ng sá»‘ vá»‘n Ä‘áº§u tÆ° cho má»—i loáº¡i Ä‘á»ƒ thá»ƒ hiá»‡n viá»‡c phÃ¢n chia Ä‘áº§u tÆ°.


```python
# Tá»•ng sá»‘ tiá»n Ä‘áº§u tÆ°
tongVonDauTu = 10000  # Giáº£ sá»­ báº¡n cÃ³ 10.000 Ä‘Ã´ la Ä‘á»ƒ Ä‘áº§u tÆ°

# Äá»‹nh nghÄ©a tá»· lá»‡ pháº§n trÄƒm Ä‘áº§u tÆ° cho má»—i loáº¡i tÃ i sáº£n
tyLe_TaiSanA = 0.4  # 40% cho TÃ i sáº£n A - cÃ³ thá»ƒ lÃ  báº¥t Ä‘á»™ng sáº£n
tyLe_TaiSanB = 0.3  # 30% cho TÃ i sáº£n B - cÃ³ thá»ƒ lÃ  cá»• phiáº¿u cÃ´ng nghá»‡
tyLe_TaiSanC = 0.3  # 30% cho TÃ i sáº£n C - cÃ³ thá»ƒ lÃ  trÃ¡i phiáº¿u hoáº·c quá»¹ Ä‘áº§u tÆ° an toÃ n

# TÃ­nh toÃ¡n sá»‘ tiá»n Ä‘áº§u tÆ° cho má»—i loáº¡i tÃ i sáº£n
dauTu_TaiSanA = tongVonDauTu * tyLe_TaiSanA
dauTu_TaiSanB = tongVonDauTu * tyLe_TaiSanB
dauTu_TaiSanC = tongVonDauTu * tyLe_TaiSanC

# In ra sá»‘ tiá»n Ä‘áº§u tÆ° cho má»—i loáº¡i tÃ i sáº£n
print(f"Sá»‘ tiá»n Ä‘áº§u tÆ° vÃ o TÃ i sáº£n A: {dauTu_TaiSanA} Ä‘Ã´ la")
print(f"Sá»‘ tiá»n Ä‘áº§u tÆ° vÃ o TÃ i sáº£n B: {dauTu_TaiSanB} Ä‘Ã´ la")
print(f"Sá»‘ tiá»n Ä‘áº§u tÆ° vÃ o TÃ i sáº£n C: {dauTu_TaiSanC} Ä‘Ã´ la")

# In ra tá»•ng quan vá» bá»©c tranh Ä‘áº§u tÆ°
print("\nBá»©c tranh Ä‘áº§u tÆ° cá»§a báº¡n bao gá»“m:")
print(f"- MÃ u xanh cá»§a sá»± an nhÃ n tá»« TÃ i sáº£n A vá»›i {tyLe_TaiSanA*100}% tá»•ng vá»‘n.")
print(f"- Sáº¯c Ä‘á» cá»§a cÆ¡ há»™i vÃ  máº¡o hiá»ƒm tá»« TÃ i sáº£n B vá»›i {tyLe_TaiSanB*100}% tá»•ng vá»‘n.")
print(f"- MÃ u vÃ ng cá»§a sá»± cÃ¢n báº±ng tá»« TÃ i sáº£n C vá»›i {tyLe_TaiSanC*100}% tá»•ng vá»‘n.")
```

> Trong ká»‹ch báº£n "PhÆ°á»£ng HoÃ ng", báº¡n nhÆ° má»™t nhÃ  tiÃªn tri Ä‘áº§y táº§m nhÃ¬n, Ä‘áº·t niá»m tin vÃ o sá»± há»“i sinh vÃ  thÄƒng hoa cá»§a thá»‹ trÆ°á»ng sau má»™t giai Ä‘oáº¡n sá»¥t giáº£m náº·ng ná». DÆ°á»›i Ä‘Ã¢y lÃ  Ä‘oáº¡n code Python thá»ƒ hiá»‡n cÃ¢u chuyá»‡n nÃ y:
