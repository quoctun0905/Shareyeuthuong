```python
import pandas as pd


# ƒê·ªçc d·ªØ li·ªáu b√°n h√†ng
df = pd.read_csv('sales_data.csv')


# Xem 5 h√†ng ƒë·∫ßu ti√™n ƒë·ªÉ hi·ªÉu d·ªØ li·ªáu
print(df.head())
```

```python
# Lo·∫°i b·ªè c√°c h√†ng c√≥ d·ªØ li·ªáu thi·∫øu
df_cleaned = df.dropna()


# ƒêi·ªÅn gi√° tr·ªã 0 v√†o c√°c √¥ tr·ªëng, n·∫øu c·∫ßn
df_filled = df.fillna(0)


# Chu·∫©n h√≥a ƒë·ªãnh d·∫°ng d·ªØ li·ªáu, v√≠ d·ª• ƒë·ªãnh d·∫°ng ng√†y th√°ng
df_cleaned['Sale_Date'] = pd.to_datetime(df_cleaned['Sale_Date'], format='%Y-%m-%d')

print('Data sau khi chu·∫©n h√≥a:')
print(df_cleaned)
```

### Ph√¢n T√≠ch ƒê∆°n Gi·∫£n


```python
# ƒê·∫øm s·ªë l∆∞·ª£ng s·∫£n ph·∫©m b√°n ƒë∆∞·ª£c theo lo·∫°i
product_counts = df_cleaned['Product_Type'].value_counts()
print("ƒê·∫øm s·ªë l∆∞·ª£ng s·∫£n ph·∫©m b√°n ƒë∆∞·ª£c theo lo·∫°i: ")
print(product_counts)

print("==================================================================")
# X√°c ƒë·ªãnh doanh s·ªë b√°n h√†ng theo th·ªùi gian
sales_over_time = df_cleaned.groupby('Sale_Date')['Quantity_Sold'].sum()
print("Doanh s·ªë b√°n h√†ng theo th·ªùi gian")
print(sales_over_time)
```

##  Ph√¢n T√≠ch Xu H∆∞·ªõng

```python
# Ph√¢n t√≠ch xu h∆∞·ªõng ti√™u d√πng theo s·∫£n ph·∫©m
trend_by_product = df_cleaned.groupby('Product_Type')['Quantity_Sold'].sum()
print("Ph√¢n t√≠ch xu h∆∞·ªõng ti√™u d√πng theo s·∫£n ph·∫©m: ")
print(trend_by_product)

print("======================================================")
# Ph√¢n t√≠ch xu h∆∞·ªõng ti√™u d√πng theo th√°ng
df_cleaned['Month'] = df_cleaned['Sale_Date'].dt.month
trend_by_month = df_cleaned.groupby('Month')['Quantity_Sold'].sum()
print("Ph√¢n t√≠ch xu h∆∞·ªõng ti√™u d√πng theo th√°ng: ")
print(trend_by_month)
```

Sau khi ƒë√£ chuy·ªÉn ƒë·ªïi d·ªØ li·ªáu th·ªùi gian, b·∫°n c√≥ th·ªÉ d·ªÖ d√†ng tr√≠ch xu·∫•t th√¥ng tin nh∆∞ nƒÉm, th√°ng, ng√†y, gi·ªù t·ª´ c·ªôt th·ªùi gian:


```python
df['year'] = df['Sale_Date'].dt.year
df['month'] = df['Sale_Date'].dt.month
df['day'] = df['Sale_Date'].dt.day

df
```

## Th·ª±c Hi·ªán Ph√¢n T√≠ch Chu·ªói Th·ªùi Gian

```python
# T√≠nh t·ªïng doanh s·ªë b√°n h√†ng theo th√°ng
monthly_sales = df.groupby(df['month'])['Quantity_Sold'].sum()


# V·∫Ω bi·ªÉu ƒë·ªì xu h∆∞·ªõng doanh s·ªë b√°n h√†ng theo th√°ng
monthly_sales.plot(kind='line')
```
