## Phần 1: Cuộc Đối Thoại Giữa Cuộc Sống và Xác Suất
Tại Sao Xác Suất Lại Quan Trọng?
Xác suất không chỉ giới hạn ở những quyết định cá nhân nhỏ nhặt; nó còn ảnh hưởng đến mọi khía cạnh của đời sống và xã hội, từ y học, kinh tế, khoa học, đến công nghệ. Mỗi khi chúng ta đứng trước lựa chọn, xác suất là người bạn đồng hành giúp chúng ta làm quen với bất định, giúp chúng ta 'đoán' tương lai dựa trên thông tin hiện tại.

## **Xác Suất trong Kinh Doanh**
###Bài Toán: Ra Mắt Sản Phẩm Mới

```python
# Định nghĩa điểm số cho từng yếu tố
market_score = 0.8
competitor_score = 0.5
production_capability_score = 0.7
marketing_strategy_score = 0.9

# Tính xác suất thành công của sản phẩm
# Điểm đối thủ cạnh tranh được điều chỉnh bởi (1 - competitor_score) vì điểm càng thấp có nghĩa là cạnh tranh càng ít
success_probability = (market_score + (1 - competitor_score) + production_capability_score + marketing_strategy_score) / 4

print(f'Xác suất thành công của sản phẩm: {success_probability}')
```

## Phần 2: Đánh giá dự án coffee

```python
# Định nghĩa điểm số và trọng số cho mỗi yếu tố (Giả định tất cả các yếu tố có trọng số như nhau trong ví dụ này)
scores = {'location': 0.8, 'competition': 0.4, 'product_quality': 0.9, 'marketing': 0.7, 'customer_service': 0.85}

# Điều chỉnh điểm cạnh tranh (1 - điểm cạnh tranh để phản ánh cạnh tranh càng ít thì càng tốt)
scores['competition'] = 1 - scores['competition']

# Tính xác suất thành công dựa trên trung bình cộng của các điểm số đã điều chỉnh
success_probability = sum(scores.values()) / len(scores)

print(f"Xác suất thành công của quán cà phê dựa trên mô hình và dữ liệu hiện tại là: {success_probability:.2f}")
```
