## Phần 1: Matplotlib/Seaborn cho dữ liệu đồ họa
Dữ Liệu Ví Dụ
Giả sử chúng ta có dữ liệu doanh số bán hàng hàng tháng cho hai sản phẩm: Cà Phê Việt và Cà Phê Mỹ, từ tháng 1 đến tháng 6 như sau:

| Tháng | Cà Phê Việt | Cà Phê Mỹ |
|-------|-------------|-----------|
| 1     | 120         | 150       |
| 2     | 135         | 160       |
| 3     | 150         | 155       |
| 4     | 145         | 170       |
| 5     | 160         | 165       |
| 6     | 175         | 180       |

Mô tả Biểu Đồ
Dựa trên dữ liệu này, chúng ta sẽ tạo một biểu đồ đường để thể hiện xu hướng doanh số bán hàng hàng tháng cho hai sản phẩm. Mỗi dòng trên biểu đồ sẽ đại diện cho một sản phẩm, với màu sắc khác nhau để dễ dàng phân biệt.


```python
import matplotlib.pyplot as plt

# Dữ liệu
thang = ['Tháng 1', 'Tháng 2', 'Tháng 3', 'Tháng 4', 'Tháng 5', 'Tháng 6']
ca_phe_viet = [120, 135, 150, 145, 160, 175]
ca_phe_my = [150, 160, 155, 170, 165, 180]


# Tạo biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(thang, ca_phe_viet, marker='o', linestyle='-', color='b', label='Cà Phê Việt')
plt.plot(thang, ca_phe_my, marker='s', linestyle='--', color='r', label='Cà Phê Mỹ')


# Đặt tiêu đề và nhãn
plt.title('Doanh Số Bán Hàng Hàng Tháng')
plt.xlabel('Tháng')
plt.ylabel('Doanh Số')
plt.legend()
plt.grid(True)


# Hiển thị biểu đồ
plt.show()
```

> Biểu đồ này giúp chúng ta dễ dàng nhận ra xu hướng tăng trưởng của doanh số bán hàng qua các tháng cho cả hai loại cà phê, cũng như so sánh hiệu suất bán hàng giữa chúng.


## Giả sử bạn có dữ liệu về doanh số bán hàng của 3 loại cà phê khác nhau trong một tháng, như sau:

| Loại cà phê | Cà Phê Việt |
|-------|-------------|
| Cà Phê Việt     | 120         |
| Cà Phê Brazil     | 135         |
| Cà Phê Mỹ    | 150         |

```python
import matplotlib.pyplot as plt

# Dữ liệu
loai_ca_phe = ['Cà Phê Việt', 'Cà Phê Brazil', 'Cà Phê Mỹ']
doanh_so = [150, 120, 180]


# Tạo biểu đồ cột
plt.figure(figsize=(8, 6))
plt.bar(loai_ca_phe, doanh_so, color=['red', 'green', 'blue'])


# Đặt tiêu đề cho biểu đồ và các trục
plt.title('Doanh Số Bán Hàng Cà Phê Trong Tháng')
plt.xlabel('Loại Cà Phê')
plt.ylabel('Doanh Số (túi)')


# Hiển thị biểu đồ
plt.show()
```

## **Giới thiệu về Seaborn:**
Seaborn là một thư viện trực quan hóa dữ liệu trong Python dựa trên Matplotlib, nhưng cung cấp một giao diện cao cấp hơn, giúp tạo ra các biểu đồ thống kê đẹp mắt một cách dễ dàng hơn. Seaborn tập trung vào việc tạo ra các biểu đồ thống kê nhanh chóng và dễ dàng, với giao diện và cách sử dụng dễ hiểu, cùng khả năng tùy chỉnh cao.


## Giả sử bạn có dữ liệu về điểm số của học sinh trong các bài kiểm tra Toán và Văn từ một lớp học, như sau:

| Học Sinh | Điểm Toán | Điểm Văn |
|----------|-----------|----------|
| HS1      | 8         | 7        |
| HS2      | 7         | 8        |
| HS3      | 9         | 6        |
| HS4      | 6         | 9        |
| HS5      | 8         | 8        |

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Tạo DataFrame từ dữ liệu
data = pd.DataFrame({
'Học Sinh': ['HS1', 'HS2', 'HS3', 'HS4', 'HS5'],
'Điểm Toán': [8, 7, 9, 6, 8],
'Điểm Văn': [7, 8, 6, 9, 8]
})

# Vẽ biểu đồ phân tán sử dụng Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Điểm Toán', y='Điểm Văn', hue='Học Sinh', palette='deep', s=100)

# Đặt tiêu đề và nhãn cho biểu đồ
plt.title('Mối Quan Hệ giữa Điểm Toán và Điểm Văn của Học Sinh')
plt.xlabel('Điểm Toán')
plt.ylabel('Điểm Văn')

plt.legend(title='Học Sinh')
plt.show()
```

Để vẽ biểu đồ xu hướng thị trường chứng khoán, chúng ta cần dữ liệu giá cổ phiếu theo thời gian. Dưới đây là một ví dụ giả định về dữ liệu giá cổ phiếu hàng ngày của một công ty trong một tuần.


## **Dùng Matplotlib:**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu
df = pd.read_csv('gia_co_phieu.csv')
df['Ngày'] = pd.to_datetime(df['Ngày'])


# Vẽ biểu đồ sử dụng Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(df['Ngày'], df['Giá Đóng Cửa'], marker='o', linestyle='-', color='b')
plt.title('Xu Hướng Giá Cổ Phiếu Theo Thời Gian - Matplotlib')
plt.xlabel('Ngày')
plt.ylabel('Giá Đóng Cửa')
plt.grid(True)
plt.show()
```

## **Sử dụng Seaborn:**

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Ngày', y='Giá Đóng Cửa', marker='o', color='r')
plt.title('Xu Hướng Giá Cổ Phiếu Theo Thời Gian - Seaborn')
plt.xlabel('Ngày')
plt.ylabel('Giá Đóng Cửa')
plt.grid(True)
plt.show()
```

#### Dưới đây là một ví dụ về dữ liệu chứng khoán giả định, nơi chúng ta sẽ xem xét không chỉ giá đóng cửa mà còn khối lượng giao dịch hàng ngày của một cổ phiếu cụ thể. Điều này sẽ cung cấp cho chúng ta cái nhìn sâu sắc hơn về hoạt động thị trường, bao gồm sự quan tâm và tính thanh khoản của cổ phiếu.


**Dùng Matplotlib:**

```python
import pandas as pd
import matplotlib.pyplot as plt


# Đọc và chuẩn bị dữ liệu
df = pd.read_csv('chung_khoan.csv')
df['Ngày'] = pd.to_datetime(df['Ngày'])


# Vẽ biểu đồ giá đóng cửa
plt.figure(figsize=(14, 7))
plt.subplot(2, 1, 1) # 2 hàng, 1 cột, vị trí thứ 1
plt.plot(df['Ngày'], df['Giá Đóng Cửa'], marker='o', linestyle='-', color='b')
plt.title('Giá Đóng Cửa Theo Thời Gian')
plt.xlabel('Ngày')
plt.ylabel('Giá Đóng Cửa')


# Vẽ biểu đồ khối lượng giao dịch
plt.subplot(2, 1, 2) # 2 hàng, 1 cột, vị trí thứ 2
plt.bar(df['Ngày'], df['Khối Lượng'], color='g')
plt.title('Khối Lượng Giao Dịch Theo Thời Gian')
plt.xlabel('Ngày')
plt.ylabel('Khối Lượng')


plt.tight_layout()
plt.show()
```

**Sử dụng Seaborn:**

```python
import seaborn as sns


# Vẽ biểu đồ giá đóng cửa với Seaborn
plt.figure(figsize=(14, 7))
plt.subplot(2, 1, 1)
sns.lineplot(data=df, x='Ngày', y='Giá Đóng Cửa', marker='o', color='r')


# Vẽ biểu đồ khối lượng giao dịch với Seaborn
plt.subplot(2, 1, 2)
sns.barplot(data=df, x='Ngày', y='Khối Lượng', color='y')


plt.tight_layout()
plt.show()
```
