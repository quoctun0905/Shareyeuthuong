### **Phần 3: ROI**

Ta biết công thức tính ROI (Return On Investment) \\


ROI = $\frac{Gain\space -\space Cost}{Cost} \cdot 100\% $

```python
def calculate_ROI(gain, cost):
  roi = (gain - cost) / cost * 100
  return roi
```

* Bạn đã thử nghiệm với những bước hướng dẫn trên và đạt được kết quả
chưa? Cảm giác của bạn thế nào, có gặp chút thách thức nào không? Có lẽ trong quá trình lập trình, bạn có thể hơi băn khoăn về bước tạo hàm với def hay không biết làm sao để sử dụng hàm đó qua câu lệnh `print(calculate_ROI(gain = 12000000, cost = 10000000))` ? Đừng lo lắng, hãy giữ bình tĩnh, mình sẽ giải cùng bạn!

* Hãy tưởng tượng `def` trong Python giống như việc bạn đang tạo ra một công thức nấu ăn. Công thức này sẽ hướng dẫn bạn làm ra một món ăn cụ thể từ những nguyên liệu bạn có. Trong lập trình, `def` giúp bạn định nghĩa một "công thức" để giải quyết một vấn đề cụ thể, mà chúng ta gọi là hàm.


```python
print(calculate_ROI(gain = 12000000, cost = 10000000))
```

### Ví dụ thực tế: Giả sử bạn muốn tính tổng tiền tiết kiệm trong hai tuần. Bạn có thể tạo một hàm để làm điều đó:

```python
def tinh_tong_tiet_kiem(tuan_1, tuan_2):
  return tuan_1 + tuan_2
```

* `def` bắt đầu công thức của bạn.
`tinh_tong_tiet_kiem` là tên bạn đặt cho công thức này, giống như bạn đặt tên cho một món ăn.
`(tuan_1, tuan_2)` chính là nguyên liệu cần thiết cho công thức, ở đây là số tiền bạn tiết kiệm được mỗi tuần.
`return tuan_1 + tuan_2` cho biết công thức sẽ làm gì với nguyên liệu: cộng chúng lại.

* Để áp dụng công thức này vào thực tế, bạn chỉ cần gọi nó và cho biết bạn đã tiết kiệm được bao nhiêu mỗi tuần:


```python
so_tien_tiet_kiem = tinh_tong_tiet_kiem(tuan_1 = 500000, tuan_2 = 600000)
print(f'Số tiền tiết kiệm được: {so_tien_tiet_kiem:,}')
```

Khi đó, `tinh_tong_tiet_kiem(500000, 600000)` chính là việc bạn áp dụng công thức nấu ăn với 500.000 đồng bạn tiết kiệm được tuần đầu và 600.000 đồng tuần thứ hai. Hàm sẽ "nấu" và "trả về" cho bạn tổng số tiền là 1.100.000 đồng. Như vậy, `def` giúp bạn tạo ra những "công thức" linh hoạt, giúp giải quyết các vấn đề cụ thể một cách gọn gàng và hiệu quả.
