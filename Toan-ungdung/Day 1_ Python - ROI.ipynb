### **Phn 3: ROI**

Ta bit cng thc tnh ROI (Return On Investment) \\


ROI = $\frac{Gain\space -\space Cost}{Cost} \cdot 100\% $

```python
def calculate_ROI(gain, cost):
  roi = (gain - cost) / cost * 100
  return roi
```

* Bn  th nghim vi nhng bc hng dn trn v t c kt qu
cha? Cm gic ca bn th no, c gp cht thch thc no khng? C l trong qu trnh lp trnh, bn c th hi bn khon v bc to hm vi def hay khng bit lm sao  s dng hm  qua cu lnh `print(calculate_ROI(gain = 12000000, cost = 10000000))` ? ng lo lng, hy gi bnh tnh, mnh s gii cng bn!

* Hy tng tng `def` trong Python ging nh vic bn ang to ra mt cng thc nu n. Cng thc ny s hng dn bn lm ra mt mn n c th t nhng nguyn liu bn c. Trong lp trnh, `def` gip bn nh ngha mt "cng thc"  gii quyt mt vn  c th, m chng ta gi l hm.


```python
print(calculate_ROI(gain = 12000000, cost = 10000000))
```

### V d thc t: Gi s bn mun tnh tng tin tit kim trong hai tun. Bn c th to mt hm  lm iu :

```python
def tinh_tong_tiet_kiem(tuan_1, tuan_2):
  return tuan_1 + tuan_2
```

* `def` bt u cng thc ca bn.
`tinh_tong_tiet_kiem` l tn bn t cho cng thc ny, ging nh bn t tn cho mt mn n.
`(tuan_1, tuan_2)` chnh l nguyn liu cn thit cho cng thc,  y l s tin bn tit kim c mi tun.
`return tuan_1 + tuan_2` cho bit cng thc s lm g vi nguyn liu: cng chng li.

*  p dng cng thc ny vo thc t, bn ch cn gi n v cho bit bn  tit kim c bao nhiu mi tun:


```python
so_tien_tiet_kiem = tinh_tong_tiet_kiem(tuan_1 = 500000, tuan_2 = 600000)
print(f'S tin tit kim c: {so_tien_tiet_kiem:,}')
```

Khi , `tinh_tong_tiet_kiem(500000, 600000)` chnh l vic bn p dng cng thc nu n vi 500.000 ng bn tit kim c tun u v 600.000 ng tun th hai. Hm s "nu" v "tr v" cho bn tng s tin l 1.100.000 ng. Nh vy, `def` gip bn to ra nhng "cng thc" linh hot, gip gii quyt cc vn  c th mt cch gn gng v hiu qu.
