```python
# Gi·∫£ ƒë·ªãnh r·∫±ng ch√∫ng ta ƒë√£ c√≥ d·ªØ li·ªáu th·ªùi ti·∫øt v√† s·ªü th√≠ch c√° nh√¢n c·ªßa ng∆∞·ªùi d√πng

# D·ªØ li·ªáu m·∫´u
thoi_tiet_hien_tai = "m∆∞a"  # C√°c t√πy ch·ªçn c√≥ th·ªÉ l√† "n·∫Øng", "m∆∞a", "l·∫°nh"
so_thich_ca_nhan = ["xem phim", "ƒëi d·∫°o", "ƒë·ªçc s√°ch"]

# H√†m g·ª£i √Ω ho·∫°t ƒë·ªông d·ª±a tr√™n th·ªùi ti·∫øt v√† s·ªü th√≠ch
def goi_y_hoat_dong(thoi_tiet, so_thich):
    hoat_dong_goi_y = []

    if "xem phim" in so_thich and thoi_tiet == "m∆∞a":
        hoat_dong_goi_y.append("xem phim t·∫°i nh√†")

    if "ƒëi d·∫°o" in so_thich and thoi_tiet == "n·∫Øng":
        hoat_dong_goi_y.append("ƒëi d·∫°o trong c√¥ng vi√™n")

    if "ƒë·ªçc s√°ch" in so_thich and (thoi_tiet == "n·∫Øng" or thoi_tiet == "m∆∞a"):
        hoat_dong_goi_y.append("th∆∞ gi√£n v·ªõi m·ªôt cu·ªën s√°ch")

    # N·∫øu kh√¥ng c√≥ ho·∫°t ƒë·ªông ph√π h·ª£p, g·ª£i √Ω m·ªôt ho·∫°t ƒë·ªông th∆∞ gi√£n kh√°c
    if not hoat_dong_goi_y:
        hoat_dong_goi_y.append("th·ª≠ m·ªôt s·ªü th√≠ch m·ªõi t·∫°i nh√†")

    return hoat_dong_goi_y

# G·ªçi h√†m v√† in ra g·ª£i √Ω
hoat_dong_duoc_goi_y = goi_y_hoat_dong(thoi_tiet_hien_tai, so_thich_ca_nhan)
print("Ho·∫°t ƒë·ªông g·ª£i √Ω cho bu·ªïi t·ªëi c·ªßa b·∫°n:")
for hoat_dong in hoat_dong_duoc_goi_y:
    print("-", hoat_dong)
```

### Ph·∫ßn 2: T√≠nh ƒëi·ªÉm h√≤a v·ªën cho s·∫£n ph·∫©m m·ªõi.
\- Sau khi kh√°m ph√° th·∫ø gi·ªõi if-else, ch√∫ng ta s·∫Ω d·∫•n th√¢n v√†o m·ªôt cu·ªôc phi√™u l∆∞u m·ªõi v·ªõi m·ªôt th√°ch th·ª©c kinh t·∫ø th√∫ v·ªã: ƒëi·ªÉm h√≤a v·ªën. H√£y c√πng nhau t√¨m hi·ªÉu xem ƒëi·ªÉm h√≤a v·ªën l√† g√¨ v√† t·∫°i sao n√≥ l·∫°i quan tr·ªçng trong vi·ªác ra quy·∫øt ƒë·ªãnh cho c√°c d·ª± √°n kinh doanh nh√©!

\- Khi b·∫Øt ƒë·∫ßu m·ªôt d·ª± √°n hay ph√°t tri·ªÉn s·∫£n ph·∫©m m·ªõi, vi·ªác t√≠nh to√°n ƒëi·ªÉm h√≤a v·ªën (break-even point) l√† m·ªôt b∆∞·ªõc quan tr·ªçng gi√∫p b·∫°n x√°c ƒë·ªãnh s·ªë l∆∞·ª£ng s·∫£n ph·∫©m c·∫ßn b√°n ƒë·ªÉ b√π ƒë·∫Øp chi ph√≠ v√† b·∫Øt ƒë·∫ßu sinh l·ªùi. D∆∞·ªõi ƒë√¢y l√† c√°ch ti·∫øp c·∫≠n ƒë·ªÉ t√≠nh ƒëi·ªÉm h√≤a v·ªën cho s·∫£n ph·∫©m m·ªõi c·ªßa b·∫°n:

* #### B∆∞·ªõc 1: X√°c ƒê·ªãnh Chi Ph√≠ C·ªë ƒê·ªãnh (Fixed Costs)
Chi ph√≠ c·ªë ƒë·ªãnh l√† nh·ªØng chi ph√≠ kh√¥ng thay ƒë·ªïi d√π s·ªë l∆∞·ª£ng s·∫£n ph·∫©m s·∫£n xu·∫•t c√≥ thay ƒë·ªïi. V√≠ d·ª•: ti·ªÅn thu√™ m·∫∑t b·∫±ng, l∆∞∆°ng nh√¢n vi√™n, chi ph√≠ qu·∫£ng c√°o.

* #### B∆∞·ªõc 2: X√°c ƒê·ªãnh Gi√° B√°n S·∫£n Ph·∫©m v√† Chi Ph√≠ Bi·∫øn ƒê·ªïi (Variable Costs)
Gi√° B√°n S·∫£n Ph·∫©m: ƒê√¢y l√† gi√° m√† b·∫°n d·ª± ƒë·ªãnh b√°n s·∫£n ph·∫©m cho kh√°ch h√†ng.
Chi Ph√≠ Bi·∫øn ƒê·ªïi: Chi ph√≠ n√†y thay ƒë·ªïi d·ª±a tr√™n s·ªë l∆∞·ª£ng s·∫£n ph·∫©m s·∫£n xu·∫•t. V√≠ d·ª•: nguy√™n li·ªáu, ti·ªÅn c√¥ng tr·ª±c ti·∫øp.

* #### B∆∞·ªõc 3: T√≠nh ƒê√≥ng G√≥p Bi√™n L·ª£i Nhu·∫≠n (Contribution Margin)
ƒê√≥ng g√≥p bi√™n l·ª£i nhu·∫≠n ƒë∆∞·ª£c t√≠nh b·∫±ng c√°ch tr·ª´ chi ph√≠ bi·∫øn ƒë·ªïi kh·ªèi gi√° b√°n s·∫£n ph·∫©m.

  > **C√¥ng th·ª©c:** ƒê√≥ng G√≥p Bi√™n L·ª£i Nhu·∫≠n = Gi√° B√°n S·∫£n Ph·∫©m ‚Äì Chi Ph√≠ Bi·∫øn ƒê·ªïi

* #### B∆∞·ªõc 4: T√≠nh ƒêi·ªÉm H√≤a V·ªën (Break-Even Point)
Cu·ªëi c√πng, b·∫°n s·∫Ω t√≠nh ƒëi·ªÉm h√≤a v·ªën b·∫±ng c√°ch chia chi ph√≠ c·ªë ƒë·ªãnh cho ƒë√≥ng g√≥p bi√™n l·ª£i nhu·∫≠n.

  > **C√¥ng th·ª©c:** ƒêi·ªÉm H√≤a V·ªën = Chi Ph√≠ C·ªë ƒê·ªãnh / ƒê√≥ng G√≥p Bi√™n L·ª£i Nhu·∫≠n
<br/><br/>

#### V√≠ D·ª• Minh H·ªça
Gi·∫£ s·ª≠ chi ph√≠ c·ªë ƒë·ªãnh cho s·∫£n ph·∫©m m·ªõi l√† 100,000 USD, gi√° b√°n d·ª± ki·∫øn l√† 50 USD/s·∫£n ph·∫©m, v√† chi ph√≠ bi·∫øn ƒë·ªïi l√† 20 USD/s·∫£n ph·∫©m.

ƒê√≥ng G√≥p Bi√™n L·ª£i Nhu·∫≠n: 50 ‚àí 20 = 30
50 ‚àí 20 = 30 USD/s·∫£n ph·∫©m
ƒêi·ªÉm H√≤a V·ªën: 100,000 √∑ 30 = 3333.33
100,000 √∑ 30 = 3333.33 s·∫£n ph·∫©m
V·∫≠y b·∫°n c·∫ßn b√°n √≠t nh·∫•t kho·∫£ng 3334 s·∫£n ph·∫©m ƒë·ªÉ h√≤a v·ªën.

#### ƒê·ªÉ t√≠nh ƒëi·ªÉm h√≤a v·ªën s·ª≠ d·ª•ng if-else, ch√∫ng ta c√≥ th·ªÉ th√™m m·ªôt s·ªë logic v√†o m√£ l·ªánh ƒë·ªÉ x·ª≠ l√Ω c√°c tr∆∞·ªùng h·ª£p kh√°c nhau, ch·∫≥ng h·∫°n nh∆∞ khi ƒë√≥ng g√≥p bi√™n l·ª£i nhu·∫≠n b·∫±ng 0 (tr√°nh chia cho 0) ho·∫∑c khi chi ph√≠ c·ªë ƒë·ªãnh b·∫±ng 0 (s·∫£n ph·∫©m kh√¥ng c·∫ßn b√°n b·∫•t k·ª≥ ƒë∆°n v·ªã n√†o ƒë·ªÉ h√≤a v·ªën). D∆∞·ªõi ƒë√¢y l√† v√≠ d·ª•:




```python
# ƒê·ªãnh nghƒ©a c√°c bi·∫øn theo v√≠ d·ª• minh h·ªça
chi_phi_co_dinh = 100000  # USD
gia_ban_san_pham = 50  # USD/s·∫£n ph·∫©m
chi_phi_bien_doi = 20  # USD/s·∫£n ph·∫©m

# T√≠nh ƒê√≥ng G√≥p Bi√™n L·ª£i Nhu·∫≠n
dong_gop_bien_loi_nhu·∫≠n = gia_ban_san_pham - chi_phi_bien_doi

# S·ª≠ d·ª•ng if-else ƒë·ªÉ x·ª≠ l√Ω c√°c tr∆∞·ªùng h·ª£p ƒë·∫∑c bi·ªát
if dong_gop_bien_loi_nhu·∫≠n <= 0:
    print("L·ªói: ƒê√≥ng g√≥p bi√™n l·ª£i nhu·∫≠n kh√¥ng h·ª£p l·ªá (<=0). Kh√¥ng th·ªÉ t√≠nh ƒëi·ªÉm h√≤a v·ªën.")
elif chi_phi_co_dinh <= 0:
    print("L∆∞u √Ω: Kh√¥ng c√≥ chi ph√≠ c·ªë ƒë·ªãnh. S·∫£n ph·∫©m kh√¥ng c·∫ßn b√°n b·∫•t k·ª≥ ƒë∆°n v·ªã n√†o ƒë·ªÉ h√≤a v·ªën.")
else:
    # T√≠nh ƒêi·ªÉm H√≤a V·ªën
    diem_hoa_von = chi_phi_co_dinh / dong_gop_bien_loi_nhu·∫≠n
    print("ƒê√≥ng G√≥p Bi√™n L·ª£i Nhu·∫≠n:", dong_gop_bien_loi_nhu·∫≠n, "USD/s·∫£n ph·∫©m")
    print("ƒêi·ªÉm H√≤a V·ªën:", round(diem_hoa_von), "s·∫£n ph·∫©m")
```

```python
# @title H√£y th·ª≠ t√≠nh ƒëi·ªÉm h√≤a v·ªën kh√°c nhau b·∫±ng c√°ch thay ƒë·ªïi th√¥ng s·ªë c√°c gi√° tr·ªã t·∫°i ƒë√¢y: {run: "auto"}
chi_phi_co_dinh = 10000 # @param {type:"integer"}
gia_ban_san_pham = 500 # @param {type:"integer"}
chi_phi_bien_doi = 20 # @param {type:"integer"}

# T√≠nh ƒê√≥ng G√≥p Bi√™n L·ª£i Nhu·∫≠n
dong_gop_bien_loi_nhu·∫≠n = gia_ban_san_pham - chi_phi_bien_doi

# S·ª≠ d·ª•ng if-else ƒë·ªÉ x·ª≠ l√Ω c√°c tr∆∞·ªùng h·ª£p ƒë·∫∑c bi·ªát
if dong_gop_bien_loi_nhu·∫≠n <= 0:
    print("L·ªói: ƒê√≥ng g√≥p bi√™n l·ª£i nhu·∫≠n kh√¥ng h·ª£p l·ªá (<=0). Kh√¥ng th·ªÉ t√≠nh ƒëi·ªÉm h√≤a v·ªën.")
elif chi_phi_co_dinh <= 0:
    print("L∆∞u √Ω: Kh√¥ng c√≥ chi ph√≠ c·ªë ƒë·ªãnh. S·∫£n ph·∫©m kh√¥ng c·∫ßn b√°n b·∫•t k·ª≥ ƒë∆°n v·ªã n√†o ƒë·ªÉ h√≤a v·ªën.")
else:
    # T√≠nh ƒêi·ªÉm H√≤a V·ªën
    diem_hoa_von = chi_phi_co_dinh / dong_gop_bien_loi_nhu·∫≠n
    print("ƒê√≥ng G√≥p Bi√™n L·ª£i Nhu·∫≠n:", dong_gop_bien_loi_nhu·∫≠n, "USD/s·∫£n ph·∫©m")
    print("ƒêi·ªÉm H√≤a V·ªën:", round(diem_hoa_von), "s·∫£n ph·∫©m")
```

#### Ch√∫ng m√¨nh ƒë√£ c√πng ƒëi qua if-else v√† c√°ch s·ª≠ d·ª•ng if-else sao cho hi·ªáu qu·∫£ v√† ·ª©ng d·ª•ng, trong h√†nh tr√¨nh h·ªçc t·∫≠p n√†y, m√¨nh th·∫•y m·ªôt s·ªë ƒëi·ªÉm c·∫ßn l∆∞u √Ω c·ªßa if-else:

üìù Note Nhanh: L∆∞u √ù Khi D√πng If-Else trong Python

üõ† C·∫©n Th·∫≠n v·ªõi Indentation: Python y√™u c·∫ßu th·ª•t ƒë·∫ßu d√≤ng ch√≠nh x√°c. Sai s√≥t ·ªü ƒë√¢y s·∫Ω d·ªÖ d·∫´n ƒë·∫øn IndentationError.

‚úÖ Ch·ªçn To√°n T·ª≠ ƒê√∫ng: Ph√¢n bi·ªát r√µ == d√πng ƒë·ªÉ so s√°nh v√† = d√πng ƒë·ªÉ g√°n gi√° tr·ªã.

üö´ Tr√°nh So S√°nh Tr·ª±c Ti·∫øp v·ªõi Boolean: Thay v√¨ if condition == True:, h√£y s·ª≠ d·ª•ng if condition:.
